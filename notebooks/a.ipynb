```python
def sol(arr):

    sum1 = 0
    sum2 = 0
    for i in range(len(arr)):
        sum1 += arr[i]
        sum2 = sum1 + arr[i+1]

        if (sum2 - sum1 == arr[i]):
            return arr[i]
```
